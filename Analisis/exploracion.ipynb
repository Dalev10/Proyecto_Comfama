</div>

En el presente dataset se encuentran los resultados obtenidos en las pruebas Saber 3°, 5° y 9° de los establecimientos educativos de Medellín para el año 2017.

* año -> Año en el que se realizaron las pruebas.
* codigo_dane -> Identificador único de cada establecimiento educativo legalmente constituido. 
* establecimiento_educativo -> Nombre de la institución o establecimiento educativo del que se recogen los resultados. 
* sector_educativo -> Tipo de establecimiento educativo (Oficial, Privado, Por contratación).
* comuna -> Comuna en la que se encuentra el establecimiento educativo.
* grado -> Grado o nivel educativo del estudiante que presento la prueba. 
* area -> Área o materia de estudio. (Matemáticas - Lenguaje)
* promedio -> Promedio o puntaje obtenido de las pruebas por estudiante (0 - 500)
* promedio_establecimiento -> Promedio o puntaje obtenido de las pruebas por institución. 
* desviacion -> Medida de de dispersión de las calificaciones obtenidas por los alumnos del establecimiento para ese grado y área.
* nivel_superior -> Porcentajes de estudiantes que se encuentran en una categoría "superior" según sus resultados.
* nivel_minimo -> Porcentajes de estudiantes que se encuentran en una categoría "mínimo" según sus resultados.
* nivel_satisfactorio -> Porcentajes de estudiantes que se encuentran en una categoría "satisfactoria" según sus resultados.
* nivel_avanzado -> Porcentajes de estudiantes que se encuentran en una categoría "avanzada" según sus resultados.
